```
import folium
import geopandas as gpd
import pandas as pd
```

```
syd_geo = gpd.read_file("/content/worldgeojson.json")
syd_geo.head(1)
```

```
syd_income = pd.read_csv("/content/wikipediadata.csv")
syd_income.head(1)
```

```
sydgeo_gdf=syd_geo.merge(syd_income,on="id")
sydgeo_gdf.head(2)
```

```
m = folium.Map([15, 74], zoom_start=6,tiles='cartodbpositron')



choropleth = folium.Choropleth(
    geo_data=syd_geo,
    name='choropleth',
    data=sydgeo_gdf,
    columns=['id', 'percentage'],
    key_on='feature.properties.id',
    fill_color='PuBuGn',
    nan_fill_color='black',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='percentage of wikipedia users worldwide',
    highlight=True,
    line_color='black'
).add_to(m)

#folium.LayerControl(collapsed=True).add_to(m)
m

suburbName =  folium.features.GeoJson(sydgeo_gdf,tooltip=folium.features.GeoJsonTooltip(fields=['id','totalviews', 'viewsperperson','percentage'],aliases=['region:','Total views:','views per person:','percentage:'],style=("background-color: white; color: #333333; font-family: arial; font-size: 12px; padding: 10px;") ))
choropleth.add_child(suburbName)
m
    

```
